## Load the data

Load data from the csv file: <i> 'iris_dataset.csv'</i>.

```python
df = pd.read_csv('iris_dataset.csv')
```

## Plot the data

For this exercise, try to cluster the iris flowers by the shape of their sepal.

<i> Use the 'sepal_length' and 'sepal_width' variables.</i>

```python
plt.scatter(df['sepal_length'],df['sepal_width'])
plt.xlabel('speal length')
plt.ylabel('sepal width')
plt.show()
```

# Clustering (unscaled data)

Separate the original data into 2 clusters.

```python
x = df.copy() #make a copy
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
clusters = df.copy()
clusters['Clusters_pred'] = kmeans.fit_predict(x)
clusters
```

```python
plt.scatter(clusters['sepal_length'],clusters['sepal_width'],c=clusters['Clusters_pred'],cmap='rainbow')
```

# Standardize the variables

Import and use the <i> method </i> function from sklearn to standardize the data.

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(df)
x_scaled
```

# Clustering (scaled data)

```python
kmeans_scaled = KMeans(2)
kmeans.fit(x_scaled)
```

```python
clusters_scaled = df.copy()
clusters_scaled['Clusters Pred'] = kmeans_scaled.fit_predict(x_scaled)
```

```python
plt.scatter(clusters_scaled['sepal_length'],clusters_scaled['sepal_width'],c= clusters_scaled['Clusters Pred'],cmap='rainbow')
```

## Take Advantage of the Elbow Method

### WCSS

```python
wcss = []
# 'cl_num' is a that keeps track the highest number of clusters we want to use the WCSS method for.
# We have it set at 10 right now, but it is completely arbitrary.
cl_num = 10
for i in range (1,cl_num):
    kmeans= KMeans(i)
    kmeans.fit(x_scaled)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
wcss
```

### The Elbow Method

```python
number_clusters = range(1,cl_num)
plt.plot(number_clusters,wcss)
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('within clusrter sum of square')
```
